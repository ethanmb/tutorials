```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (12,8)

#read in data

df = pd.read_csv("/Users/ethanbrown/Desktop/data-analyst-portfolio/dataCorrelation_py/movies.csv")
```

```python
#print out data to see if read
df.head()
```

```python
#check for missing data
for col in df.columns:
    pctMissing = np.mean(df[col].isnull())
    print("{} - {}%".format(col, pctMissing))
```

```python
#dropped all rows with missing data and recheck for missing data
df = df.dropna()
for col in df.columns:
    pctMissing = np.mean(df[col].isnull())
    print("{} - {}%".format(col, pctMissing))
```

```python
# Data types for columns
df.dtypes
```

```python
#changed data type of budget, gross, votes, and runtime to clean up look as all were .0
df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
df["votes"] = df["votes"].astype("int64")
df["runtime"] = df["runtime"].astype("int64")

df
```

```python
#create new column with correct release year, extracts year from the string
df["yearCorrect"] = df["released"].astype(str).str.extract('(?P<year>\d{4})').astype(int)

df
```

```python
df.sort_values(by=["gross"],inplace=False, ascending=False)
```

```python
pd.set_option("display.max_rows", 25)
```

```python
# Drop duplicates, of which there are none
df.drop_duplicates()
```

```python
#plot budget vs gross using matlib
plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs. Gross Earnings")
plt.xlabel("Budget")
plt.ylabel("Gross Earnings")

plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color": "grey"}, line_kws={"color":"blue"})
```

```python
df.corr()
```

```python
#create and diplay correlation matrix
correlationMatrix = df.corr()
sns.heatmap(correlationMatrix, annot=True, linewidths=1, cmap = "Greens")
plt.title("Correlation Matrix")

plt.show()
```

```python
#"numerize" data so it can be correlated
dfNumerize = df

for colName in dfNumerize.columns:
    if(dfNumerize[colName].dtype == "object"):
        dfNumerize[colName] = dfNumerize[colName].astype("category")
        dfNumerize[colName] = dfNumerize[colName].cat.codes
        
dfNumerize
```

```python
#create and diplay correlation matrix
correlationMatrix = dfNumerize.corr()
sns.heatmap(correlationMatrix, annot=True, linewidths=1, cmap = "Greens")
plt.title("Correlation Matrix")

plt.show()
```

```python
dfNumerize.corr()
```

```python
corrMat = dfNumerize.corr()
corrPairs = corrMat.unstack()

sortedPairs = corrPairs.sort_values()
sortedPairs
```

```python
highCorr = sortedPairs[(sortedPairs > 0.5)]

highCorr
```
